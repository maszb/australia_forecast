```python
import folium
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_villes=pd.read_csv('AUSLONLAT.csv',sep=';')
```

```python
df_villes
```

```python
villesref=pd.read_csv("VilleRef.csv",sep=";")
```

```python
villesref
```

```python
df_villes_ref={}
for idx,ligne in villesref.iterrows():
    df_villes_ref[ligne['Ville']]=ligne['Radar']
df_villes_ref
```

```python
df_villes_previ={}
for ville in df_villes['city']:
    df_villes_previ[ville]={'pluie': 0,'radar':0}
df_villes_previ
```

```python
for idx,ligne in villesref.iterrows():
    if ligne['Radar']=='isole':
        df_villes_previ[ligne['Ville']]['radar']=1
    else:
        df_villes_previ[ligne['Ville']]['radar']=0
sorted(df_villes_previ.items())
```

```python
df_villes_previ['Darwin']['pluie']=1
df_villes_previ['Katherine']['pluie']=1
df_villes_previ['Cairns']['pluie']=1
df_villes_previ['Townsville']['pluie']=1
```

```python
df_villes_previ
```

```python
#En rouge les villes radar
#En bleu les villes normales
#A l'intérieur soleil = beau temps, goutte eau =pluie
```

```python
from folium.features import CustomIcon
m=folium.Map(location=[df_villes[df_villes['city']=='Sydney']['lat'], df_villes[df_villes['city']=='Sydney']['lng']], zoom_start=5)
url="C:/Users/rblv6364/Documents/Important/MachineLearning/Projet/{}".format
urlimage=url("IconSun.png")
urlimageradar=url("IconSun2.png")
for index, row in df_villes.iterrows():
    if df_villes_previ[row['city']]['radar']==1:
        colradar='red'
        pos=folium.features.CustomIcon(urlimageradar,icon_size=(30,30))
    else:
        colradar='blue'
        pos=folium.features.CustomIcon(urlimage,icon_size=(30,30))
    if df_villes_previ[row['city']]['pluie']==1:
        Libelle=" il pleuvra"
        popup="<h2>"+row['city']+"</h2><p>demain"+ Libelle+"</p>"
        marker=folium.Marker(location=[row['lat'], row['lng']],popup=popup,tooltip=row['city'],icon=folium.Icon(icon='tint',icon_color='blue',color=colradar))
        marker.add_to(m)
    else:
        Libelle=" il ne pleuvra pas"
        popup="<h2>"+row['city']+"</h2><p>demain"+ Libelle+"</p>"
        marker=folium.Marker(location=[row['lat'], row['lng']],popup=popup,tooltip=row['city'],icon=pos)
        marker.add_to(m)
m
```

```python
#Traçage des villes et mise en avant des villes radar en rouge
m=folium.Map(location=[df_villes[df_villes['city']=='Sydney']['lat'], df_villes[df_villes['city']=='Sydney']['lng']], zoom_start=5)
tooltip = "Click Here to see the prediction"
for index, row in df_villes.iterrows():
    if df_villes_previ[row['city']]['radar']==1:
        colradar='red'
    else:
        colradar='blue'
    marker = folium.Marker(location=[row['lat'], row['lng']],popup=popup,tooltip=row['city'],icon=folium.Icon(icon='map-marker',color=colradar))
    marker.add_to(m)
m
```

```python
#Traçage des villes avec prévision
#icon=folium.Icon(icon=IconSun,icon_color=couleur,color=colradar)
m=folium.Map(location=[df_villes[df_villes['city']=='Sydney']['lat'], df_villes[df_villes['city']=='Sydney']['lng']], zoom_start=5)
tooltip = "Click Here to see the prediction"
for index, row in df_villes.iterrows():
    if df_villes_previ[row['city']]['radar']==1:
        colradar='red'
    else:
        colradar='blue'
    if df_villes_previ[row['city']]['pluie']==1:
        Libelle=" il pleuvra"
        couleur='blue'
        icon='tint'
    else:
        Libelle=" il ne pleuvra pas"   
        couleur='yellow'
        icon='certificate'
    popup="<h2>"+row['city']+"</h2><p>demain"+ Libelle+"</p>"
    marker = folium.Marker(location=[row['lat'], row['lng']],popup=popup,tooltip=row['city'],icon=folium.Icon(icon=icon,icon_color=couleur,color=colradar))
    marker.add_to(m)
m
```
